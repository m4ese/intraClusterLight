```python
import matplotlib.patches as pat
import matplotlib as mpl
from matplotlib.ticker import ScalarFormatter

mpl.rcParams['mathtext.fontset'] = 'custom'

# changes the style of plot 
# sns.set(rc={'figure.figsize':(30,18)})
sns.set_style('ticks',{"xtick.major.size":8,
"ytick.major.size":8})
fig , ax = plt.subplots()

# adjusts the ticks
ax.tick_params(axis='both', which = 'both', length = 8, labelsize = 30, bottom = True, top = True, 
               left = True, right = True, direction = 'in', pad = 15)
ax.ticklabel_format(style='plain', axis = 'both')

#this sets the width of the lines
sns.set_context(rc={"lines.linewidth": 5})

# this makes the actual plot
sns_plot = sns.lineplot(x="r / r_200", y="$f^*$", hue='Bahcall and Kulier, 2014',
            palette={"$6 \leq N \leq 8$": "r", "$9 \leq N \leq 25$": "g", "$26 \leq N \leq 220$": 'b',
                     "30-20": 'k',"20-20": 'c','30-30': 'y', '0.001-0.19': 'k'},
           markers=["^", "o", 's', '<','>','^','s'], markersize = 24, data=all_colorssf, dashes = False, 
                style = 'Bahcall and Kulier, 2014', legend = 'full')

# sets the axes scales and limits
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlim(0.2,50)
ax.set_ylim(0.003,0.1)


ax.xaxis.set_major_formatter(ScalarFormatter())
ax.yaxis.set_major_formatter(ScalarFormatter())

# legend size
plt.legend(prop={'size': 30},loc = 'upper left', fontsize = 12)

# axes labels
ax.set_title(r"Stellar Fraction vs. Radius", fontsize = 45, fontname = "Arial", 
             fontweight='bold', y = 1.03)
ax.set_xlabel(r"$\mathbf{r}$ $\mathbf{/}$ $\mathbf{r_{200}}$", fontsize = 45, fontname = "Arial")

ax.set_ylabel(r"$\mathbf{f*}$ = $\mathbf{M_*}$ / $\mathbf{M_{200}}$ ($\mathbf{r}$ / $\mathbf{r_{200}}$)",
              fontsize = 45, fontname = "Arial", fontweight = 'bold')

# plots vertical bar for error bar
plt.axvspan(xmin = 1.96, xmax = 2.0, ymin= 0.203, ymax = 0.373, facecolor = 'g', alpha = 0.75)

# and the horizontal bars
plt.axhspan(ymin = 0.008649 + 0.00256 - 0.0001, ymax = 0.008649 + 0.00256, xmin = 0.4025, xmax = 0.430, 
            color = 'g', alpha = 0.75)
plt.axhspan(ymin = 0.008649 - 0.00256, ymax = 0.008649 - 0.00256 + 0.00005, xmin = 0.4025, xmax = 0.430, 
            color = 'g', alpha = 0.75)

# Band and text labelling the band
plt.axhspan(ymin = 0.00882, ymax = 0.01078, xmin = 0, xmax = 1, color = 'c',alpha = 0.35)
plt.text(x = 20.0, y = 0.012, s ='f* Cosmic', fontsize = 40, fontweight = 'bold')

plt.show()
fig.savefig('stellarfrac.png', dpi = 100)
```
