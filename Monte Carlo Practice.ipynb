```python
#import libraries
import random
import matplotlib.pyplot as plt
```

```python
#create function for simulating die roll
#the die can take values from 1 to 100. 
# If the number is between 1 and 51, the house wins (False)
#If the number is between 52 and 100, the player wins (True)

def rolldice():
    dice = random.randint(1,100)
    
    if dice <= 51:
        return False
    elif dice > 51 and dice <=100:
        return True
```

# Here is our specific situation

```python
# create a function that simulates the bets
# 3 arguments:
#1: funds = total money the player has
#2: wager = number that the player bets each time
#3: plays = number of times the player plays

def play(money, wager, plays):
    # empty lists to keep track of the play number, available funds, and final funds
    play_num = []
    Funds = []
    
    #play number 1
    play = 1
    
    #if number of plays is less than the number of games we're going to play:
    
    while play <= plays:
        #if we win:
        #add the money
        if rolldice():
            money += wager
        
            #add the play
            play_num.append(play)
        
            #add the new fund amount
            Funds.append(money)
        #if the house wins
        else:
            #subtract the money
            money -= wager
            
            #add the play
            play_num.append(play)
            
            #add the funds amount
            Funds.append(money)
        
        play += 1
        
    #plot of funds
    plt.plot(play_num,Funds)
    Final_funds = Funds[-1]
    return(Final_funds)
```

# This is the Monte Carlo Part here!

```python

#call the above fxn to simulate the plays and calculate the remaining
#funds of the player after all the bets

#list to calculate final funds
ROI = []
#initial funds, wager, and plays
BET = 10000
WAGER = 100
PLAYS = 1000000

for x in range(1,101):
    ending_fund = play(BET, WAGER, PLAYS)
    ROI.append(ending_fund)

plt.ylabel('Player Money in $')
plt.xlabel('Number of bets')
plt.xlim([0.5, PLAYS+0.5])
plt.show()

# calculate the avererage money the player ends with
average = sum(ROI) / len(ROI)

#and print it
print("The player starts with ${} and ends with ${}".format(BET,average))

    
```
